```python
#always disconnect the program first then the recording
#note to self: make another version for peak dbfs

from pydub import AudioSegment

f = open("test.csv","r")
txt = f.read()
lines = txt.split("\n")
lines.pop(len(lines)-1) #theres a very annoying empty line at the bottom of the csv i'm too lazy to remove in the earlier code
for line in range(len(lines)):
  lines[line] = list(map(int,lines[line].split(",")))

f.close()

if len(lines[-1]) == 1:
  lines.pop(len(lines)-1)


path = "/content/drive/MyDrive/TAC 2025(Tian Cheng & Gavriel)/gqrx_20250405_133424_96300000.wav"
recording = AudioSegment.from_file(path, "wav")[:lines[-1][1]+400]
print(f"The length of the recording is {lines[-1][1]/1000} seconds long")
first_sound_list = []
second_sound_list = []
prev = 0
for i in range(len(lines)):
  first_sound_list.append(recording[prev:lines[i][0]])
  second_sound_list.append(recording[lines[i][0]:lines[i][1]])
  prev = lines[i][1]

sound1export = first_sound_list[0]
sound2export = second_sound_list[0]
f = open("loudness.csv", 'w')
f_peak = open("loudness_peak.csv","w")
f.write("1st Frequency RMS, 2nd Frequency RMS, 1st Frequency Peak dBFS, 2nd Frequency Peak dBFS, 1st Frequency Peak Power, 2nd Frequency Peak Power, 1st Frequency Peak dBFS, 2nd Frequency Peak dBFS\n")
for i in range(len(first_sound_list)):
  if i >= len(second_sound_list):
    break


  else:
    f.write(str(first_sound_list[i].rms)+","+str(second_sound_list[i].rms)+","+str(first_sound_list[i].dBFS)+","+str(second_sound_list[i].dBFS)+","+str(first_sound_list[i].max)+","+str(second_sound_list[i].max)+","+str(first_sound_list[i].max_dBFS)+","+str(second_sound_list[i].max_dBFS)+"\n")
    f_peak.write("\n")
    sound1export += first_sound_list[i]
    sound2export += second_sound_list[i]
f.close()
sound1export.export("first_sound.wav", format="wav")
sound2export.export("second_sound.wav", format="wav")

```
